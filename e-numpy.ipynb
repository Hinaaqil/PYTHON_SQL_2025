Broadcasting rules
describes how munpy treats arrays with different shapes during arithmetic operations

```python
#same shapes arrays can be added
#different shapes arrays can also be added how? broadcasting concept

#smaller array is broadcasted across larger array so that they have compatible shapes
#smaller array will repeat/broadcast.

#numpy handles it internally we dont need to instruct it. we learning broadcasting just to know the concept what internally happens.

#       1 1 1           0 1 2
#       1 1 1     +
#       1 1 1

#the small array will repeat  

#       1 1 1           0 1 2
#       1 1 1     +     0 1 2  repeat
#       1 1 1           0 1 2  repeat

#solution

#        1 2 3
#        1 2 3
#        1 2 3



```

```python
#broadcasting rules
#3 rules

#1 make the two arrays have the same number of dimensions 
#if one is 2d and one 1d first bring both in same dimension make 1d to 2d
#add new dimension with size 1 to the head of array with smaller dimension
#    (3,) 1d     (3,2) 2d  
#    (1,3)

#    (3,3,3)   (3,)         
#    (1,1,3)

#2 rule 2.....
#make each dimension of the two arrays the same

#3 rule 3.....
#if there is dimension where size is not 1 in either of the two arrays it cant be broadcasted and an error is raised.
# pick the one which is 1 and strech it equal to the bigger

# (3,3)    (3,)       1,3   rule 1
#                     3,3   rule 2   strech the 1 equal to the bigger one.


#in short first make dimension equal by adding 1 to head of smaller one.
#then strech dimension of smaller array equal to bigger one.

# (5,3)  (3,1)     already both 2d same so rule 1 already satisfied
# strech the 1 equal to the other array.   so  (5,3)  (3,3)     
#  1 is important in any one of the two arrays otherwise error
```
