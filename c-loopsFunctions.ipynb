if elif else

```python
x=10
y=20
if x<y:
    print("x is less than y")
else:
    print("x is greater than y")
```

```python
#elif use
#if previous condition false try this condition
a=33
b=33
if b>a:
    print("b is greater than a")
elif a==b:
    print("a and b are equal")
```

```python
#nested if
x=41
if x>10:   #outer if statement
    print("Above ten")
    if x>20:    #inner if st.
        print("and also above 20")
    else:
        print("but not above 20")
```

```python
marks = 62
attendance=84
if marks >60:
    if attendance >=80:
        print("student promoted")
    else:
        print("Student not promoted")
```

```python
# if elif
#1st true others will not be executed
#if you want all statements to get exucted use if if if..
price=500
discount=0
if price<100:
     discount=price*0.9 #10 percent discount so discounted price will be reaming 90% of the price
elif price>=100:
     discount=price*0.8 #20% discount
elif price>=1000:
     discount=price*0.5  #50% discount
print("Discounted price:", discount)
```

```python
#reusable code
#defining functions
#calling functions
#parameters/arguments

#def function_name(arguments):
  #body
  #return



def my_sum(num1,num2):
    result=num1+num2
    return result
#function call
my_sum(4,5)         #two arguments function
```

```python
def my_sum2(my_iterable):
    total=0
    for i in my_iterable:
        total+=i
    return total
    
#function call
a=[1,2,3,4,5]
my_sum2(a)
```

```python
#2 positional arguments --- are read in order bcz first argument assigned to 1st para. 2nd to second...
                              #follow the parameters positions
def greet(name,message):
    print("Hello",name)
    print(message)

#call
greet("jack","whats going on?")          #asAccordingToPositions--greet(name,message)--calledByPosition
```

```python
#keyword arguments----we can not only pass arguments by position but also by name

#called by keyword/names
greet(message="Howdy?",name="Jill")
    #here order doesnt matter bcz we used names
```

```python
#arbitrary arguments
#there can be any # of variables

def myfun(*vrbl):
    for avg in vrbl:
        print(avg)
```

```python
#Recursion  -- calling same function within that function but with a different argument.
#factorial of 5----  5*factorial(4)

def factorial(n):
    if(n==0 or n==1):
        return 1
    else:
        return n* factorial(n-1)        #In this line we called the function with diff arg.
    
#call
factorial(5)
```

Enumerate function

```python
#when you are itrating list loop and want it to tell you which # element is currently going on
# 2 ways -- temp variable + 1  or another way is enumerate function use helps access index of values also

#method#01 usin temp variable index=0
marks=[12,56,32,98,12,45,1,4]
index=0
for mark in marks:
    print(mark)
    if index==3:
        print("Harry owesome!")
    index +=1
```

```python
#method#02-- Using enumerate function---- values with their index also can be accessed
for index,mark in enumerate(marks):
    print(mark)
    if index==3:
        print("Harry, awesome!")
```
