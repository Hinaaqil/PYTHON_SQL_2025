Pandas
Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.

```python
#dataframe
import pandas as pd

df=pd.DataFrame({'population':[35,63,80,20],'GDP':[312,600,350,425],'surface area':[18373,18373,18363,19374],'HDI':[10,12,19,15],
                 'continent':['abe','dez','ghi','jkl']},columns=['population','GDP','surface area','continent'])   #columns needed otherwise it will create index
df
```

```python
#setting custom index
df.index=['canada','france','germany','italy']
```

```python
#for 2d arrays       #we will only use tabular data so just remember this way of fetching 2d array or table. no need learning 3d or aboves
#in numpy array we used 
#fetch 1st 3rd 4th 6th rows
#a[[0,2,3,5]]

#and used foloowing to fetch columns
#a[:,[0,2,3]]  #the second part is columns whilw the 1st part means : all rows

#used following for boolean indexing
#a[(a>50)&(a%2==0)]

#fetch 2nd and 3rd row and 2nd and 3rd column
#a2[1:,1:3]
```

```python
###indexing 
#only rows retrieve   loc   iloc

print(df.loc['canada'])

print()

print(df.iloc[-1])

print()

print(df.iloc[[0,1,-1]])  #multiple rows/records retrieve

print()
```

```python
#column retrieve   -- no need to use loc iloc bcz thats index/related to records
print(df[['population','GDP']])   #only columns retrieve

print()

print(df.loc['france':'italy','population'])  #rows plus one column ret.

print()

print(df.loc['france':'italy',['population','GDP']])  #rows plus columns retrieve
```

```python
##### iloc
print(df.iloc[1:3],[0,3])   #1st one rows   2nd one columns. rows plus columns ret.

print()

print(df.iloc[1:3,1:3])   #rows and columns ret. by slicing
```

Series
A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

```python
#import pandas
#!pip install pandas

import pandas as pd
#import numpy as np
```

```python
#series ---- like python list or numpy array

population =pd.Series([35,63,80,60])    #S capital. Series
population
```

```python
#name this series
population.name='G7 population in millions'
population
```

```python
population.dtype
```

```python
type(population)
```

```python
#population.values #it will convert it into numpyarray
print(type(population.values))

population.index
```

```python
population.index=['canada','france','germany','italy',]   #use , at the end
```

```python
#initiating + renaming + indexing     in one step
population=pd.Series({'canada':35,'france':63,'italy':26,'japan':22},name='G7 population')
population
```

```python
#another way using list
population=pd.Series([35,63,27,48],index=['canada','france','italy','japan'],name='G7 population')
population
```

```python
#calling by index
pd.Series(population,index=['france','japan'])
```

```python
print(population[0])   #at 0 index is canada so population 35

print(population['canada'])  #35 population
```
