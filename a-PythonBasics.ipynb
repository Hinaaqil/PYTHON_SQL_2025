```python
#doc strings
#docstrings are not comments python gives it a special treatment
#these are string literals that appear right after the definition of a function, method,class or module
```

```python
def square(n):
    "Takes in a number n,returns the square of n"
    print(n**2)

square(5)  #function call
```

```python
print(square.__doc__)  # returns the docstring text
```
